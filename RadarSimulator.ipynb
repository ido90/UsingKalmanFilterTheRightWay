#### General plotter

```python
def show_detections(rr, targets):
    axs = utils.Axes(4,4, axsize=(4,3))

    radar.show_detections(rr, ax=axs[0])
    radar.show_detections(rr, 't', ax=axs[1])
    radar.show_detections(rr, 't', ax=axs[2], show_fa=False)
    radar.show_detections_vs_targets(targets[0], ax=axs[3])

    plt.tight_layout()
    
    return axs
```
